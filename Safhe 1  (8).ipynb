# مشخصات زمانی و فرکانسی سیگنال‌ها و سیستم‌ها

## 6-0 مقدمه  
توصیف یک سیستم LTI بر حسب پاسخ فرکانسی جایگزینی برای توصیف حوزه زمان سیستم بر حسب کانولوشن است. در تحلیل سیستم‌های LTI غالباً استفاده از حوزه فرکانسی مناسب‌تر است زیرا معادلات دیفرانسیل و تفاضلی و عمل کانولوشن در حوزه فرکانس به عملیات جبری تبدیل می‌شوند. همچنین مفاهیمی چون فرکانس‌گزینی را در حوزه فرکانس می‌توان راحت تصور کرد. البته در طراحی سیستم معمولاً باید هم ملاحظات حوزه فرکانس منظور شود و هم ملاحظات حوزه زمان. 

## 6-1 نمایش اندازه - فاز تبدیل فوریه  
تبدیل فوریه در حالت کلی مختلط است و همان‌طور که قبلاً گفته‌ایم می‌توان آن را با بخش‌های حقیقی و موهومی یا اندازه و فازش نشان داد. نمایش اندازه - فاز تبدیل فوریه پیوسته در زمان $$  X(j\omega) $$   عبارت است از:

$$ 
X(j\omega) = |X(j\omega)| e^{j\angle X(j\omega)} 
$$


به نحوی مشابه نمایش اندازه - فاز تبدیل فوریه گسسته در زمان  X(jw)  عبارت است از:

$$(2-6)$$
$$ X(j\omega) = |X(j\omega)| e^{j\angle X(j\omega)} $$

در بحث زیر راجع به نمایش‌های اندازه - فاز توجه عمده خود را روی حالت پیوسته در زمان متمرکز می‌کنیم. نکات اصلی را به حالت گسسته در زمان نیز می‌توان اعمال کرد.

از معادله ترکیب تبدیل فوریه یعنی معادله (4-8) چنین بر می‌آید که می‌توان  X(jw)  را تجزیه‌ی سیگنال  X(t) به مجموع نمایی‌های مختلط با فرکانس‌های مختلف دانست. در حقیقت، چنان که در بخش (7-3-4) گفتیم،  |X(jw)|^2  را می‌توان طیف چگالی انرژی  X(t)  دانست. یعنی:

$$ |X(j\omega)|^2 \frac{d\omega}{2\pi} $$

را می‌توان انرژی واقع در باند فرکانسی بی‌نهایت کوچک بین w و w+dw فرض کرد. پس اندازه‌ی  |X(jw)|  محتوای فرکانسی سیگنال را نشان می‌دهد، یعنی  |X(jw)|  اطلاعات مربوط به اندازه نسبی نمایی‌های مختلط تشکیل‌دهنده‌ی  X(t)  را در بر دارد. مثلاً اگر در خارج یک باند فرکانسی کوچک حول صفر داشته باشیم  |X(jw)| = 0 ،  X(t)  تنها نوسان‌های فرکانس پایین دارد.

ولی زاویه فاز   <X(jw)  بر اندازه‌ی مولفه‌های فرکانسی اثری ندارد، بلکه اطلاعات مربوط به فاز نسبی این نمایی‌ها را نشان می‌دهد. رابطه‌ی فاز موجود در  <X(jw)  اثر زیادی بر طبیعت سیگنال  X(t)  دارد و نوعاً اطلاعات زیادی در مورد سیگنال دارد. در واقع، بسته به تابع فاز می‌توانیم سیگنال‌هایی با ظاهر متفاوت به دست آوریم، حتی اگر تابع اندازه تغییر نکند. برای نمونه، دوباره مثال شکل 3-3 را در نظر بگیرید. در اینجا کشتی با جمع آثار سه موج روبروست که هر کدام با یک سیگنال سینوسی مدل شده‌اند. اگر این سه سینوسی دامنه‌ی ثابتی داشته باشند، بسته به فاز نسبی‌شان می‌توانند دامنه‌ی بسیار بزرگ یا کوچکی داشته باشند. پس فازها برای کشتی کاملاً مهم‌اند.

به عنوان نمونه‌ی دیگری از اثر فاز، سیگنال زیر را در نظر بگیرید:

$$(6-3)$$
$$ x(t) = 1 + \frac{1}{2} \cos(2\pi t + \varphi_1) + \cos(4\pi t + \varphi_2) + \frac{2}{3} \cos(6\pi t + \varphi_3) $$

شکل 4-3 این سیگنال را به ازای $$ \varphi_1 = \varphi_2 = \varphi_3 = 0 $$ نشان می‌دهد. در شکل 1-6، این  X(t)  را به ازای چند فاز مخالف برای سینوس‌ها نشان داده‌ایم. همان‌طور که از شکل برمی‌آید، می‌بینیم که سیگنال حاصل برحسب مقادیر مختلف فاز نسبی چقدر می‌تواند متفاوت باشد.

به طور کلی، تغییر تابع فاز  X(jw)  به تغییر مشخصات حوزه‌ی زمان سیگنال  X(t)  می‌انجامد. در بعضی موارد، اعوجاج فاز مهم است و در بعضی موارد نه. مثلاً یکی از خواص سیستم شنوایی انسان عدم حساسیت به فاز است. اگر تبدیل فوریه‌ی سیگنال یک صدای تنها (مثلاً یک مصوت) اعوجاج فاز داشته باشد، ولی اندازه‌اش تغییر نکند، قابلیت فهم آن تغییر زیادی نمی‌کند، گرچه شکل موج آن در حوزه‌ی زمان می‌تواند شکلی کاملاً دگرگون به خود بگیرد. هرچند اعوجاج ملایم فاز، نظیر اعوجاج تک‌صداها، قابلیت تشخیص را از بین نمی‌برد، ولی اعوجاج فاز شدید گفتار، مسلماً قابلیت تشخیص آن‌ها را از بین می‌برد.

به عنوان مثالی بارز فرض کنید  x(t)  سیگنال ضبط‌شده‌ی بیان یک جمله است. اگر نوار وارونه پخش شود، سیگنال  X(-t)  به دست می‌آید و اثر متناظر در حوزه‌ی فرکانس، منفی شدن فاز است:

$$ F\{x(-t)\} = X(-j\omega) = |X(j\omega)| e^{-j\angle X(j\omega)} $$
