```python
import numpy as np
import matplotlib.pyplot as plt

#   
t = np.linspace(0, 10, 1000)  #  0  10  1000 

#   ( )
carrier = np.cos(10 * np.pi * t)

#   ( )
message = 1 + 0.5 * np.cos(2 * np.pi * t)

#   (AM)
modulated_signal = message * carrier

#   
plt.figure(figsize=(8, 4))
plt.plot(t, modulated_signal, 'k')  #   ('k')
plt.xlabel("t")
plt.ylabel("Amplitude")
plt.title("() -  ")
plt.grid()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#   
t = np.linspace(0, 10, 1000)

#     
carrier = np.cos(10 * np.pi * t)  #  

#   (Envelope)
message = 1 + 0.4 * np.cos(2 * np.pi * t)  #      ()

#   (AM)
modulated_signal = message * carrier

#        
modulated_signal[500] += 1.5  #     

#   
plt.figure(figsize=(8, 4))
plt.plot(t, modulated_signal, 'k')  #   ('k')
plt.xlabel("t")
plt.ylabel("Amplitude")
plt.title("() -  AM      ")
plt.grid()

plt.show()
```
