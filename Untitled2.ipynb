```python
import numpy as np
import matplotlib.pyplot as plt

# تعریف متغیر زمان
t = np.linspace(0, 10, 1000)  # از 0 تا 10 با 1000 نمونه

# سیگنال حامل (فرکانس بالا)
carrier = np.cos(10 * np.pi * t)

# سیگنال پیام (فرکانس پایین)
message = 1 + 0.5 * np.cos(2 * np.pi * t)

# مدولاسیون دامنه (AM)
modulated_signal = message * carrier

# رسم سیگنال مدوله‌شده
plt.figure(figsize=(8, 4))
plt.plot(t, modulated_signal, 'k')  # رنگ مشکی ('k')
plt.xlabel("t")
plt.ylabel("Amplitude")
plt.title("(الف) - سیگنال مدوله‌شده")
plt.grid()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# تعریف متغیر زمان
t = np.linspace(0, 10, 1000)

# سیگنال حامل (فرکانس بالا)
carrier = np.cos(10 * np.pi * t)

# سیگنال پیام (فرکانس پایین)
message = 1 + 0.5 * np.cos(2 * np.pi * t)

# مدولاسیون دامنه (AM)
modulated_signal = message * carrier

# اضافه کردن یک نویز ضربه‌ای در یک نقطه تصادفی
modulated_signal[450] += 1  # تغییر مقدار در نقطه‌ای خاص

# رسم سیگنال مدوله‌شده
plt.figure(figsize=(8, 4))
plt.plot(t, modulated_signal, 'k')  # رنگ مشکی ('k')
plt.xlabel("t")
plt.ylabel("Amplitude")
plt.title("(ب) - سیگنال مدوله‌شده با نویز ضربه‌ای")
plt.grid()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# تعریف متغیر زمان
t = np.linspace(0, 10, 1000)

# سیگنال حامل (فرکانس بالا)
carrier = np.cos(20 * np.pi * t)  # افزایش فرکانس نسبت به قبل

# سیگنال پیام (فرکانس پایین‌تر و ضعیف‌تر)
message = 1 + 0.2 * np.cos(2 * np.pi * t)  # دامنه کمتر

# مدولاسیون دامنه (AM)
modulated_signal = message * carrier

# اضافه کردن نویز ضربه‌ای در یک نقطه خاص
modulated_signal[500] += 1  # ایجاد یک پرش در وسط سیگنال

# رسم سیگنال مدوله‌شده
plt.figure(figsize=(8, 4))
plt.plot(t, modulated_signal, 'k')  # رنگ مشکی ('k')
plt.xlabel("t")
plt.ylabel("Amplitude")
plt.title("(ج) - سیگنال مدوله‌شده با فرکانس بالا و نویز ضربه‌ای")
plt.grid()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# تعریف محور زمان
t = np.linspace(0, 10, 1000)

# سیگنال حامل با فرکانس متوسط
carrier = np.cos(10 * np.pi * t)  # فرکانس متوسط

# سیگنال پیام (Envelope)
message = 1 + 0.4 * np.cos(2 * np.pi * t)  # پاکت موج قوی‌تر نسبت به (ج)

# مدولاسیون دامنه (AM)
modulated_signal = message * carrier

# اضافه کردن نویز ضربه‌ای در یک نقطه خاص
modulated_signal[500] += 1.5  # ایجاد پرش در وسط سیگنال

# رسم سیگنال مدوله‌شده
plt.figure(figsize=(8, 4))
plt.plot(t, modulated_signal, 'k')  # رنگ مشکی ('k')
plt.xlabel("t")
plt.ylabel("Amplitude")
plt.title("(د) - سیگنال AM با فرکانس متوسط و نویز ضربه‌ای")
plt.grid()

plt.show()
```
